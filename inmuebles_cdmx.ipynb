###¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
datos['Valor'].mean()
```

Eso el promedio de todo. Necesito agrupar para sacar por grupo

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

se incluyo el numeric_only porque sino daba error si se dejaba con el paréntesis vacio

```python
datos.groupby('Tipo')['Valor'].mean(numeric_only=True)
```

Ahora se puede omitir el numeric porque son numeros. A continuación se ordenará en orden ascendente, en formato dataframeordenará para verlo más facil

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Ahora almacenaré esa consulta en un avariable

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='green')
```

##**Tratamiento datos nulos**

En este caso se imndica que se les asigne el valor cero, pero eso no es un estándar, depende del caso a caso.

##**Remover Registros inconsistentes**

```python
df.query('Valor==0 | Condominio==0')
```

Cada uno de esos registros tiene indices, y esos usaré para remover

```python
df.query('Valor==0 | Condominio==0').index
```

```python
df_remover = df.query('Valor==0 | Condominio==0').index
```

```python
df.drop(df_remover,axis=0, inplace=True)
```

##**CREANDO FILTROS**

###**FILTRAR: 1 Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.**


```python
seleccion1 = df['Habitaciones'] == 1
seleccion1
```

```python
seleccion2 = df['Valor'] < 4200
seleccion2
```

Una manera es esta:

```python
filtro1 = (seleccion1) & (seleccion2)
df1 = df[filtro1]
df1
```

Otra manera es directa

```python
df[seleccion1 & seleccion2]
```

```python
df1 = df[seleccion1 & seleccion2]
df1
```
