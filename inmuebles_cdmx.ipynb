#**DEMANDAS DE ML**

###¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
datos['Tipo']
```

```python
datos['Valor'].mean()
```

Eso el promedio de todo. Necesito agrupar para sacar por grupo

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

se incluyo el numeric_only porque sino daba error si se dejaba con el paréntesis vacio

```python
datos.groupby('Tipo')['Valor'].mean(numeric_only=True)
```

Ahora se puede omitir el numeric porque son numeros. A continuación se ordenará en orden ascendente, en formato dataframeordenará para verlo más facil

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Ahora almacenaré esa consulta en un avariable

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='green')
```

Ahora iremos a por la segunda parte, que es obtener solo lo de lo residencial, pero lo aremos por complemento

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales= ['Casa Comercial', 'Conjunto Comercial/Sala', 'Departamento en Hotel',
                        'Edificio Completo', 'Galpón/Depósito/Almacén', 'Hotel', 'Industria',
                        'Posada/Chalé', 'Tienda en Centro Comercial','Tienda/Salón', 'Cochera/Estacionamiento']
```

Aqui tuve que tipear el arreglo que contiene los comerciales, aunque no coincido 100% con el profesor, y se define como variable

Ahora se consultará si dentro de la variable recien definida se encuentrrea lo que buscamos en el Data Set (el que obtuvoimos como unique

```python
datos.query('@inmuebles_comerciales in Tipo')
```

Obtuve un Data Frame con los comerciales. Ahora hare la consulta en negativo para los residenciales

```python
datos.query('@inmuebles_comerciales not in Tipo')
```

Y ahora reciclaromos el metodo data frame para el grafico y la estadística reemplazando Datos por df directamente. Antes definimos la variable df


```python
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh',figsize=(12,8),color='red')
```
